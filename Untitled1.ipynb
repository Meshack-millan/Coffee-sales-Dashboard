Q1. Show the number of confirmed, Deaths and Recovered Cases in each region

```python
df.groupby("Region")[["Confirmed","Deaths","Recovered"]].sum().head()
```

Q2. Remove all the records where confirmed caes is Less than 10

```python
df.drop(df[df["Confirmed"]<10].index,inplace=True)
```

In which region, maximum number of confirmed cases were recorded

```python
df.groupby("Region")["Confirmed"].sum().sort_values(ascending=False).head()
```

Q4. In which Region, minimum number of Death Cases were recorded

```python
df.groupby("Region")["Deaths"].sum().sort_values(ascending=True).head()
```

Q.5) How many confirmed, Deaths & Recovered cases were reported from India till 29 April 2020?

```python
df[(df["Region"]=="India") & (df["Date"]<="2020-04-29")][["Confirmed","Deaths","Recovered"]].sum()
```

Q.6) Sort the entire data wrt No. of Confirmed cases in ascending order.

```python
df.sort_values(by=["Confirmed"],ascending=True).head()
```

Q6-B) Sort the entire data with number of Recovered cases in descending order

```python
df.sort_values(by=["Recovered"],ascending=False).head()
```
