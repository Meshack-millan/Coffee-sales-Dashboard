```python
df[df["Weather"]=="Clear"]["Weather"].count()
```

```python
df.groupby("Weather").get_group("Clear")["Weather"].count()
```

```python
df[df["Wind Speed_km/h"]==4]["Wind Speed_km/h"].count()
```
